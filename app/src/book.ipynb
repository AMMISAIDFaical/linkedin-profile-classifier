```python
import json
import pandas as pd

# Keep only relevant fields from 'profile details'
keep_keys = ['about', 'experience', 'header', 'headline']

def extract_relevant_details(profile_json):
    if pd.isna(profile_json):  # skip empty rows
        return None
    try:
        data = json.loads(profile_json) if isinstance(profile_json, str) else profile_json
        return {k: data.get(k) for k in keep_keys if k in data}
    except Exception as e:
        print(f"Error parsing row: {e}")
        return None

# Apply extraction
df['profile details cleaned'] = df['profile details'].apply(extract_relevant_details)

# Optional: preview the cleaned output for first rows
print(df['profile details cleaned'].head().to_list())
```

```python
# Expand the cleaned dicts into separate columns
df_expanded = df.join(pd.json_normalize(df['profile details cleaned']))
```
